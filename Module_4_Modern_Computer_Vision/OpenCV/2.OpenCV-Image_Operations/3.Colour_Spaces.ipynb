```python
# Our Setup, Import Libraries and Create our Imshow Function
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Load our input image
image = cv2.imread('images/flowers.jpg')

# Use cv2.split to get each colour space separately
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
# Each colour space on it's on will look like a grayscale as it lacks the other colour channels
imshow("Blue Channel Only", B)
```

```python
import numpy as np

# Let's create a matrix of zeros
# with dimensions of the image h x w
zeros = np.zeros(image.shape[:2], dtype = "uint8")

imshow("Red", cv2.merge([zeros, zeros, R]))
imshow("Green", cv2.merge([zeros, G, zeros]))
imshow("Blue", cv2.merge([B, zeros, zeros]))
```

```python
image = cv2.imread('images/flowers.jpg')

# OpenCV's 'split' function splits the image into each colour index
B, G, R = cv2.split(image)

# Let's re-make the original image
merged = cv2.merge([B, G, R])
imshow("Merged", merged)
```

```python
# Let's amplify the blue colour
merged = cv2.merge([B+100, G, R])
imshow("Blue Boost", merged)
```

### **The HSV Colour Space** ###

```python
# Reload our image
image = cv2.imread('images/flowers.jpg')
```

```python
# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
imshow('HSV', hsv_image)
```

```python
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
plt.show()
```

```python
# Switching back to viewing the RGB representation
imshow("Hue", hsv_image[:, :, 0])
imshow("Saturation", hsv_image[:, :, 1])
imshow("Value", hsv_image[:, :, 2])
```
